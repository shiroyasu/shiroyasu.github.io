# 複素関数論 (4) #

## 複素数の関数 ##

### 複素関数 ###

- 複素数の値をとる変数 $z$ と $w$ がある（複素変数）．
$z$ に対して，$w$ が唯一つ決まるとき，$w$ は $z$ の関数であるといい，$w=f(z)$ と表す．
- $z=x+iy$ とおけば， $w$ は2つの独立変数 $x, y$ に関する関数 $w=f(x, y)$ と考えられる．
- さらに，$f(x, y)$ も複素数なので， 実部と虚部に分けることができる； $w=f(x, y)=u(x,y)+i\,v(x,y)$
- このように， 複素変数の複素数値関数 $w=f(z)$ は， 2つの2変数実数値関数 $u(x,y),\ v(x,y)$ の組と考えることができる．

### 複素関数の極限 ###

-  関数$w=f(z)$について， $z$が数（点）$a$に近づく（ただし，$a$に一致することなく）につれて， その経路の如何に関わらず（どんな近づけ方をしても）， $f(z)$ の値が１つの定まった数$b$に限りなく近づくとする．
このとき，
$$\lim_{z\rightarrow a}f(z)=b\quad\mbox{または}\quad f(z)\rightarrow b\ (z\rightarrow a)$$
と書き表す．
- $z$ が数（点）$a$ に近づく（ただし， $a$ に一致することなく）につれて， その経路の如何に関わらず（どんな近づけ方をしても）， $|f(z)|$ が限りなく大きくなるとする．このとき，
$$\lim_{z\rightarrow a}f(z)=\infty\quad\mbox{または}\quad f(z)\rightarrow \infty\ (z\rightarrow a)$$
と書き表す．
- $|z|$ が限りなく大きくなるように， $z$を動かすとき，その経路の如何に関わらず（どんな動かし方をしても）， $f(z)$の値が１つの定まった数$b$に限りなく近づくとする．このとき，
$$\lim_{z\rightarrow \infty}f(z)=b\quad\mbox{または}\quad f(z)\rightarrow b\ (z\rightarrow \infty)$$
と書き表す．

### 複素関数の連続性 ###

- 関数$w=f(z)$の定義域内の点 $a$ について, $\lim_{z\rightarrow a}f(z)=f(a)$ が成り立つとき， 「$w=f(z)$は$a$で<b>連続</b>である」という．
- 定義域内の領域$D$内のすべての点で連続であるとき， 「$w=f(z)$は$D$で連続である」という．
- $f(z), g(z)$が連続ならば， $f(z)\pm g(z)$， $f(z)\,g(z)$， $\frac{f(z)}{g(z)}$も連続である．
- $f(z)=f(x+iy)=u(x, y)+i\,v(x, y)$ が連続であるための必要十分条件は， 平面$\mathbb{R}^2$上の関数 $u, v$ が連続であることと同値である．

### 複素関数の可視化 ###

```
JupyterImageResolution = 72;
```

#### (I) 2つの実2変数関数（実部と虚部）として可視化 ####

$z=x+yi$とおくと, $f(z)=u(x,y)+i\,v(x,y)$と表すことができ, 実部$u(x,y)$と虚部$v(x,y)$は, それぞれ（実）2変数関数である.

(i) $f(z)=z^2$

```
GraphicsRow[{
  Plot3D[Re[(x + I*y)^2], {x, -1, 1}, {y, -1, 1}],
  Plot3D[Im[(x + I*y)^2], {x, -1, 1}, {y, -1, 1}]
  }]
```

(ii) $f(z)=1/z$

```
GraphicsRow[{
  Plot3D[Re[1/(x + I*y)], {x, -1, 1}, {y, -1, 1}],
  Plot3D[Im[1/(x + I*y)], {x, -1, 1}, {y, -1, 1}]
  }]
```

(iii) $f(z)=\dfrac{z}{\bar{z}}$

```
GraphicsRow[{
  Plot3D[Re[(x + I*y)/(x - I*y)], {x, -1, 1}, {y, -1, 1}],
  Plot3D[Im[(x + I*y)/(x - I*y)], {x, -1, 1}, {y, -1, 1}]
  }]
```

この関数は$z=0$で連続ではない（p.130<b> 例題3</b>を参照）.

#### (II) 2つの実2変数関数（絶対値と偏角）として可視化 ####

$z=x+yi$に対し, $f(z)$の絶対値$|f(z)|$および, 偏角$\arg(f(z))$を対応させる関数は，いずれも（実）2変数関数である.

```
GraphicsRow[{
  Plot3D[Abs[(x + I*y)^2], {x, -1, 1}, {y, -1, 1}],
  Plot3D[Arg[(x + I*y)^2], {x, -1, 1}, {y, -1, 1}]
  }]
```

```
GraphicsRow[{
  Plot3D[Abs[1/(x + I*y)], {x, -1, 1}, {y, -1, 1}],
  Plot3D[Arg[1/(x + I*y)], {x, -1, 1}, {y, -1, 1}]
  }]
```

```
GraphicsRow[{
  Plot3D[Abs[(x + I*y)/(x - I*y)], {x, -1, 1}, {y, -1, 1}],
  Plot3D[Arg[(x + I*y)/(x - I*y)], {x, -1, 1}, {y, -1, 1}]
  }]
```

$w=\dfrac{z}{\bar{z}}$とすると, $|w|=1$であることがわかる.

#### (III) 絶対値の曲面を偏角の値で色付け ####

$z=x+yi$に対し, $f(z)$の絶対値$|f(z)|$を対応させる（実）2変数関数を偏角$\arg(f(z))$の値に応じて連続的に色付けする.

```
Show[
 ParametricPlot[{r*Cos[t], r*Sin[t]}, {r, 2/3, 1}, {t, -Pi, Pi}, 
  Mesh -> None, Axes -> False, Frame -> False, BoundaryStyle -> None, 
  ColorFunction -> Function[{x, y}, Hue[Rescale[Arg[x + I*y], {0, 2*Pi}]]], 
  ColorFunctionScaling -> False],
 Graphics[Arrow[{{-1.2, 0}, {1.2, 0}}]],
 Graphics[Arrow[{{0, -1.2}, {0, 1.2}}]], PlotRange -> All
 ]
```

```
Plot3D[Abs[(x + I*y)^2], {x, -1, 1}, {y, -1, 1}, 
 ColorFunction -> 
  Function[{x, y, z}, Hue[Rescale[Arg[(x + I*y)^2], {0, 2*Pi}]]], 
 ColorFunctionScaling -> False]
```

```
Plot3D[Abs[1/(x + I*y)], {x, -1, 1}, {y, -1, 1}, 
 ColorFunction -> 
  Function[{x, y, z}, Hue[Rescale[Arg[1/(x + I*y)], {0, 2*Pi}]]], 
 ColorFunctionScaling -> False]
```

```
Plot3D[Abs[(x + I*y)/(x - I*y)], {x, -1, 1}, {y, -1, 1}, 
 ColorFunction -> 
  Function[{x, y, z}, Hue[Rescale[Arg[(x + I*y)/(x - I*y)], {0, 2*Pi}]]], 
 ColorFunctionScaling -> False]
```

#### (IV) $z$-平面の曲線の像を$w$-平面に描画 ####

$z$-平面のメッシュ（軸に平行な直線族）が, $f(z)$によってどのような曲線族に移るのか, 可視化する.

```
u=1/2;
v=1/2;
GraphicsRow[{
  Show[
   ParametricPlot[Through[{Re, Im}[(x + I*y)]], {x, -1, 1}, {y, -1, 1}, 
    Mesh -> None],
   Graphics[{PointSize[0.02], Point[{u, v}]}],
   Graphics[{Red, Line[{{u, -1}, {u, 1}}], Blue,Line[{{-1, v}, {1, v}}]}]],
  Show[
   ParametricPlot[Through[{Re, Im}[(x + I*y)^2]], {x, -1, 1}, {y, -1, 1},Mesh -> None],
   Graphics[{PointSize[0.04],Point[Through[{Re, Im}[(u + I*v)^2]]]}],
   ParametricPlot[Through[{Re, Im}[(u + I*t)^2]], {t, -1, 1},PlotStyle -> {Thickness[0.008], Red}, Mesh -> None],
   ParametricPlot[Through[{Re, Im}[(t + I*v)^2]], {t, -1, 1},PlotStyle -> {Thickness[0.008], Blue}, Mesh -> None]]
  }]
```

```
u=1/2;
v=1/2;
GraphicsRow[{
  Show[
   ParametricPlot[Through[{Re, Im}[(x + I*y)]], {x, -1, 1}, {y, -1, 1},Mesh -> None],
   Graphics[{PointSize[0.04], Point[{u, v}], Red,Line[{{u, -1}, {u, 1}}], Blue, Line[{{-1, v}, {1, v}}]}],
   PlotRange -> {{-1, 1}, {-1, 1}}],
  Show[
   ParametricPlot[Through[{Re, Im}[1/(x + I*y)]], {x, -1, 1}, {y, -1, 1}, Mesh -> None],
   Graphics[{PointSize[0.04], Point[Through[{Re, Im}[1/(u + I*v)]]]}],
   ParametricPlot[Through[{Re, Im}[1/(u + I*t)]], {t, -1, 1},PlotStyle -> {Thickness[0.008], Red}],
   ParametricPlot[Through[{Re, Im}[1/(t + I*v)]], {t, -1, 1},PlotStyle -> {Thickness[0.008], Blue}], 
   PlotRange -> {{-4, 4}, {-4, 4}}]
  }]
```
