# 複素関数論 (7) #

## 複素関数の線積公式 ##

ここでの主題は, 「<b>単一閉曲線</b>を積分経路とする複素積分」である.

### 単一閉曲線 ###

曲線 $C : z(t), a\leqq t\leqq b$ に対し,

- $C$ が<b>単一曲線</b>である：自分自身と交わることのない曲線. つまり, $t_1\ne t_2$ ならば $z(t_1)\ne z(t_2)$ である.
- $C$ が<b>閉曲線</b>である：始点と終点が同じ曲線. つまり, $z(a)= z(b)$.
- 単一曲線かつ閉曲線を<b>単一閉曲線</b>（または<b>単純閉曲線</b>）という.

```
JupyterImageResolution = 72;
```

例） 円周は単一閉曲線である.

```
ParametricPlot[2*{Cos[t],Sin[t]},{t,0,2*Pi}]
```

一般に複素数平面において, 中心が $\alpha\in \mathbb{C}$ で 半径が $r>0$ の円周上の点 $z$ は
$$
|z-\alpha|=r
$$
を満たす. これが<b>円周の方程式</b>である.

- $z-\alpha$ の絶対値が $r$ だから, $z-\alpha$を極表示すると
$$
z-\alpha=r(\cos t+i\sin t)
$$
と書ける.
- 指数関数を用いると, $r(\cos t+i\sin t)=r e^{ti}$ であるため,
$$z=\alpha+r e^{ti}$$と書ける.
これが, 円周のパラメータ表示 $z(t)$ である（ただし, $0\leqq t\leqq 2\pi$）.
- 中心 $\alpha$ を $\alpha=a+bi$ とおけば, 実部と虚部は
$$
z=(r\cos t-a)+i(r\sin t-b)
$$
と書ける.

例） レムニスケートは閉曲線であるが, 単一曲線ではない（原点で交差している）.

```
ParametricPlot[Sqrt[Cos[2*t]]*{Cos[t],Sin[t]}/2,{t,-Pi,Pi
                                                }]
```

### コーシーの定理 ###

単一閉曲線 $C$ とその内部で, 関数 $f(z)$ は正則であるとする.
このとき,
$$
\int_C f(z)\,dz=0
$$
である.

例１） $f(z)=z-\alpha$, $C : z(t)=\alpha+e^{ti}, 0\leqq t\leqq 2\pi$

$C$ は $\alpha$ を中心とする半径 $1$ の円周である.
関数 $f(z)$ は複素数平面全体で正則であるから, $C$ とその内部でも正則である.
よって, コーシーの定理から, 積分値は$0$である.

実際に, 定義に従って計算してみる.
曲線 $C$ のパラメータ表示は $z(t)=\alpha +e^{ti}$ であるから, $z'(t)=i\,e^{ti}$.
よって,
$$
\int_C f(z)\,dz
=\int^{2\pi}_0 \{(\alpha +e^{ti})-\alpha\}\,i\,e^{ti}\,dt
=i\int^{2\pi}_0 e^{2ti}\,dt
=\frac{1}{2}\left[e^{2ti}\right]^{2\pi}_0
=\frac{1}{2}(e^{4\pi i}-e^0)=0
$$

例２） $f(z)=\dfrac{1}{z-\alpha}$, $C : z(t)=\alpha+e^{ti}, 0\leqq t\leqq 2\pi$

$C$ は $\alpha$ を中心とする半径 $1$ の円周である.
関数 $f(z)$ は複素数平面から1点$z=\alpha$を除いた領域で正則であるが, この点は $C$ の内部にあるので,
コーシーの定理の仮定を満たさない.

曲線 $C$ のパラメータ表示は $z(t)=\alpha +e^{ti}$ であるから, $z'(t)=i\,e^{ti}$.
よって,
$$
\int_C f(z)\,dz
=\int^{2\pi}_0 \frac{1}{(\alpha +e^{ti})-\alpha}\,i\,e^{ti}\,dt
=i\int^{2\pi}_0 \frac{1}{e^{ti}}\,e^{ti}\,dt
=i\int^{2\pi}_0 dt
=i\left[t\right]^{2\pi}_0
=2\pi i
$$
となる.

例３） $f(z)=\dfrac{1}{z-\alpha}$, $C : z(t)=\alpha+2e^{ti}, 0\leqq t\leqq 2\pi$

$C$ は $\alpha$ を中心とする半径 $2$ の円周である.
関数 $f(z)$ は複素数平面から1点$z=\alpha$を除いた領域で正則であるが, この点は $C$ の内部にあるので,
コーシーの定理の仮定を満たさない.

曲線 $C$ のパラメータ表示は $z(t)=\alpha +2e^{ti}$ であるから, $z'(t)=2i\,e^{ti}$.
よって,
$$
\int_C f(z)\,dz
=\int^{2\pi}_0 \frac{1}{(\alpha +2e^{ti})-\alpha}\,2i\,e^{ti}\,dt
=2i\int^{2\pi}_0 \frac{1}{2e^{ti}}\,e^{ti}\,dt
=i\int^{2\pi}_0 dt
=i\left[t\right]^{2\pi}_0
=2\pi i
$$
となり, 例２）の結果と同じになった.

一般に次の定理が成り立つ.

#### 定理. ####

関数 $f(z)$ は領域 $D$ で正則であるとする.
２つの単一閉曲線 $C_1$ と $C_2$ は次の2条件を満たす;
<ol>
    <li>一方は他方の内部に含まれる.</li>
    <li>2つの閉曲線で囲まれた（挟まれた）領域は $D$ に含まれる.</li>
</ol>
このとき,
$$
\int_{C_1}f(z)\,dz=\int_{C_2}f(z)\,dz
$$
が成り立つ.

#### 系. ####

関数 $f(z)$ は領域 $D$ で正則であるとする.
単一閉曲線 $C$ の内部に, $k$個の単一閉曲線 $C_1,\ldots,C_k$ があり（これらは交点を持たない）, $C$と$C_1,\ldots,C_k$で囲まれた（挟まれた）領域は $D$ に含まれるとする.
このとき,
$$
\int_{C}f(z)\,dz=\int_{C_1}f(z)\,dz+\int_{C_2}f(z)\,dz+\cdots+\int_{C_k}f(z)\,dz
$$
が成り立つ.

### コーシーの積分表示（その１） ###

$f(z)$ を領域 $D$ で正則な関数とする.
$C$ を領域 $D$ 内の単一閉曲線 とし, $C$ の内部も $D$ に含まれているとする.

このとき,
$$
f(a)=\frac{1}{2\pi i}\int_C \frac{f(z)}{z-a}dz
$$
が成り立つ.

上の式は
$$
\int_C \frac{f(z)}{z-a}dz=2\pi i\,f(a)
$$
とみることができる. つまり, 積分値が1点での関数値で決定される.

### コーシーの積分表示（その２：グルサの公式） ###

$f(z)$ を領域 $D$ で正則な関数とする.
$C$ を領域 $D$ 内の単一閉曲線 とし, $C$ の内部も $D$ に含まれているとする.

このとき,
$$
f^{(n)}(a)=\frac{n!}{2\pi i}\int_C \frac{f(z)}{(z-a)^{n+1}}dz
$$
が成り立つ.

#### 系. ####

正則な関数 $f(z)$ は何回でも微分できる.
また, $f(z)$ の導関数 $f^{(n)}(z)$ は $f(z)$ が正則な領域で正則である.
